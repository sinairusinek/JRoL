# Example: Daniel Sanders' correspondance over time

This notebook was written by Dror Guldin, Avigail Friedland and Anne K in the framework of the Potsdam DH-Jewish Hackathon 2022 in order to provide an example of a possible visualization of letter relationship over time.

**Input**

The notebook takes as input a CSV with the following fields:
1. Sender - Name of the person sending the letter
2. Recipient  - Name of the person receiving the letter
3. Date - Date in which the letter was sent

**Output**

The notebook outputs an interactive plot showing each letter over a timeline, grouped by the recipient.

```python
import plotly.express as px
import pandas as pd
```

```python
dfSenders= pd.read_csv('sanders_letters - SandersRecipient.csv')
```

```python
# x and y given as array_like objects
# scatterplot all recepients
figAgain = px.scatter(x=dfSenders['date'], y=dfSenders['recipient'], labels={
                     "x": "Time",
                     "y": "Recipients of Sanders's letters"
                     
                 }, title="Sanders' recipients of letters over time")

figAgain.show()
```
